```python
%load_ext autoreload
%autoreload
from inspect import getmembers, isfunction
from hlee_notClass_multiplexed_lee22e import *
from hlee_utils import *

################
#  unpack()    # 
################
def unpack(ct,ha,wa,ca,ka,ta):
    tsize = ha*wa*ka**2
    ch  = []
    for channel in range(ta*ka**2):
        r = channel%(ka**2)
        idx_start_channel = ta*(r%ka)+int(np.floor(r/ka))*ta*ka*wa +int(np.floor(channel/(ka**2)))
        mat=[]
        for i in range(ha):
            row = []
            for j in range(wa):
                idx = idx_start_channel+i*wa*ta*ka**2+j*ka*ta
                row.append(ct[idx])
            mat.append(row)
        ch.append(mat)
    return np.array(ch)


####################
#  unpack_sum()    #
####################
def unpack_sum(ct,ha,wa,ca,ka,ta,ci):
    tsize = ha*wa*ka**2*ci
    ch  = []
    for channel in range(ka**2*ta):
        r = channel%(ka**2)
        idx_start_channel = ta*(r%ka)+int(np.floor(r/ka))*ta*ka*wa +int(np.floor(channel/(ka**2)))
        mat=[]
        for i in range(ha):
            row = []
            for j in range(wa):
                idx = idx_start_channel+i*wa*ta*ka**2+j*ka*ta+ci
                sum=0
                for s in range(ci):
                    sum+=ct[idx+s]
                #row.append(ct[idx])
                row.append(sum)
            mat.append(row)
        ch.append(mat)
    return np.array(ch)

####################
#  unpack_sum3()    #
####################
def unpack_sum3(ct,ha,wa,ca,ka,ta,ci):
    tsize = ha*wa*ka**2*ci
    ch  = []
    for channel in range(ka**2*ta):
        r = channel%(ka**2)
        idx_start_channel = ta*(r%ka)+int(np.floor(r/ka))*ta*ka*wa +int(np.floor(channel/(ka**2)))
        mat=[]
        for i in range(ha):
            row = []
            for j in range(wa):
                idx = idx_start_channel+i*wa*ta*ka**2+j*ka*ta+ci
                sum=0
                for s in range(ci):
                    sum+=ct[idx+s]
                #row.append(ct[idx])
                row.append(sum)
            mat.append(row)
        ch.append(mat)
    return np.array(ch)
```

# 1st Trial.
- A : $((row*100+col*10+ch*1)./1000 )*1000^{ch}$
- U : leave A unchanged after convolution

```python
#############################
#  Dimensions, Params       #
#############################
fh,fw = 3,3
hi,wi,ci,ki,ti = 3,3,4,1,4 ;ho,wo,co,ko,to = 3,3,4,1,4
kernels= [fh,fw];ins  = [hi,wi,ci,ki,ti];outs = [ho,wo,co,ko,to]
nslots = 2**14
#############################
#  kernels that do nothing  #
############################
U = np.zeros([co,ci,fh,fw])
for l in range(co):
    for k in range(ci):
        for i in range(fh):
            for j in range(fw):
                if i==1 and j==1:
                    U[l,k,i,j]=1
U = get_channel_last(U)
U_ch1st = get_channel_first(U)
#############################
#  Dummy Small image        #
#############################
A = np.zeros([hi,wi,ci])
for i in range(hi):
    for j in range(wi):
        for k in range(ci):
            if k != (ci-1):
                A[i,j,k]=((i+1)*100+(j+1)*10+k+1)/1000
                if k>0:
                    A[i,j,k]*=1000**k
A_ch1st = get_channel_first(A)
ct_a = MultPack(A,ki,ti)
ct_d = MultConv(ct_a,U,ins,outs,kernels,nslots=2**14)
out = unpack_noC(ct_d,ho,wo,co,ko,to,ci)
out2 = unpack(ct_d,ho,wo,co,ko,to)
l = np.zeros([hi,1])
combined1 = np.hstack([out[0],l,out[1],l,out[2],l,out[3]])
combined2 = np.hstack([out2[0],l,out2[1],l,out2[2],l,out2[3]])
imshow_with_value(combined1,0)
imshow_with_value(combined2,0)
```
