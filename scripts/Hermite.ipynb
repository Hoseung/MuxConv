```python
import math 
import numpy as np
import matplotlib.pyplot as plt
```

```python
def HNrelu(x, n=2):
    H0 = 1
    H1 = x
    H2 = x**2 - 1
    #if n==3: H3 = 8*x**3 - 12*x

    # Normalized Hermite polynomials
    h0 = H0
    h1 = H1
    h2 = H2/math.sqrt(2)
    #if n==3: h3 = H3/math.sqrt(6)

    # Coefficients of Hermite approximation of ReLU
    f0 = 1/math.sqrt(2*math.pi)
    f1 = 1/2
    f2 = 1/math.sqrt(2*math.pi*2) # try 1/() if it doesn't work
    # double factorial
    # 3!! = 3 * 1 
    # 4!! = 4 * 2
    # 5!! = 5 * 3 * 1
    # 6!! = 6 * 4 * 2
    # n!! = n * (n-2) * (n-4) * ... * 1 if n is odd
    # n!! = n * (n-2) * (n-4) * ... * 2 if n is even
    # (n-3)!! = 1 or 1*3 or 1*3*5. maybe 1*3. 
    return f0*h0 + f1*h1 + f2*h2 
def relu(x):
    return np.maximum(0, x)
```

```python
x = np.linspace(-4, 4, 100)
plt.plot(x, relu(x), label='ReLU')
plt.plot(x, HNrelu(x), label='HNReLU')
```
